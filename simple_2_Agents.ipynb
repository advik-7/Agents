```python
from transformers import pipeline

researcher_llm = pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B", max_length=200)
writer_llm = pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B", max_length=500)

class ResearcherAgent:
    def __init__(self, llm):
        self.llm = llm

    def search_and_summarize(self, query):
        prompt = f"Search for information on {query} and summarize the key points."
        summary = self.llm(prompt)[0]['generated_text']
        return summary

class WriterAgent:
    def __init__(self, llm):
        self.llm = llm

    def write_speech(self, research_summary):
        prompt = f"Write a keynote speech based on the following research: {research_summary}"
        speech = self.llm(prompt)[0]['generated_text']
        return speech

researcher = ResearcherAgent(researcher_llm)
writer = WriterAgent(writer_llm)

topic = "Quantum Computing"
research_summary = researcher.search_and_summarize(topic)

speech = writer.write_speech(research_summary)

print("Research Summary:")
print(research_summary)
print("\nGenerated Speech:")
print(speech)
```
